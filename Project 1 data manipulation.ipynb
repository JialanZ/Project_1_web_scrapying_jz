```python
#import modules
import pandas as pd
import numpy as np
import re
```

```python
#import movie_ratings_us.csv
!cat movie_ratings_us.csv
```

```python
#import movie_ratings_us.csv as pd data frame
mr_us=pd.read_csv('movie_ratings_us.csv')
mr_us.head(10)
```

```python
#dimension of the data frame
mr_us.shape
```

```python
#number of missing values for each variable
print('the number of missing genre is :', np.sum(mr_us.genre.isnull()))
print('the number of missing gross is :', np.sum(mr_us.gross.isnull()))
print('the number of missing imdb rating is :', np.sum(mr_us.imdb_rating.isnull()))
print('the number of missing meta rating is :', np.sum(mr_us.meta_rating.isnull()))
print('the number of missing movie name is :', np.sum(mr_us.movie_name.isnull()))
print('the number of missing movie rated is :', np.sum(mr_us.movie_rated.isnull()))
print('the number of missing vote number is :', np.sum(mr_us.n_of_votes.isnull()))
print('='*50)
print('the number of movie with imdb rating is : ', mr_us.shape[0] - np.sum(mr_us.imdb_rating.isnull()))
print('the number of movie with meta rating is : ', mr_us.shape[0] - np.sum(mr_us.meta_rating.isnull()))
```

```python
#convert all words into lower case: extract the column first, then apply map function:
mr_us.genre=mr_us.genre.map(lambda x: x.lower())
mr_us.movie_name=mr_us.movie_name.map(lambda x: x.lower())
mr_us.head(10)
```

```python
#create movie_name_match for later matching with movie_rating_douban.csv
name_for_match_us=mr_us.movie_name.map(lambda x: ' '.join(re.findall('[a-z]+',x)))
mr_us['movie_name_for_match']=name_for_match_us
mr_us.head()
#when I use mr_us.movie_name_for_match=movie_name_for_match it doesn't work! need to figure out why
```

```python
#missing value exist because some movie names are weird! no characters at all, ignore them, they may trigger problems when merging
mr_us.replace('', np.nan,inplace=True)
mr_us.dropna(subset=['movie_name_for_match'], inplace=True)
mr_us_f=mr_us
#mr_us_f=mr_us[mr_us.movie_name_for_match.notnull()]    #dont understand why this one doesnt work!
print(np.sum(mr_us_f.movie_name_for_match.isnull()))
mr_us_f.head(70)
```

```python
#export the final version data frame
#mr_us_f.to_csv('movie_ratings_us_f.csv')
```

```python
#remove observations with missing imdb ratings: pd.notnull() returns true if it is not missing value
mr_imdb=mr_us_f[mr_us_f.imdb_rating.notnull()]
print(mr_imdb.shape)
print('the number of missing imdb rating is :', np.sum(mr_imdb.imdb_rating.isnull()))
```

```python
#remove observations with missing meta ratings
mr_meta=mr_us_f[mr_us_f.meta_rating.notnull()]
print(mr_meta.shape)
print('the number of missing meta rating is :', np.sum(mr_meta.meta_rating.isnull()))
```

```python
#remove observations with either missing ratings:
mr_both=mr_us_f[mr_us_f.imdb_rating.notnull() & mr_us_f.meta_rating.notnull()]
print(mr_both.shape)
print('the number of missing imdb rating is :', np.sum(mr_both.imdb_rating.isnull()))
print('the number of missing meta rating is :', np.sum(mr_both.meta_rating.isnull()))
```

```python
mr_both.to_csv('mr_both_no_stack.csv')
```

```python
#stack the movie with multi genres
#use reset_index to make movie_name a normal column
mr_us_f=mr_us_f.set_index('movie_name').genre.str.split(', ',expand=True).stack().reset_index()
```

```python
mr_us_f.head()
```

```python
#somehow the column name of genre becomes 0, change it to GENRE
mr_us_f.rename(columns={0:'GENRE'},inplace=True)
mr_us_f.head()
```

```python
#merge the stacked data frame with the original one
mr_us_new=pd.merge(mr_us, mr_us_f, how='right', on='movie_name')
```

```python
mr_us_new.head()
```

```python
#drop the unnecessary columns (genre, movie_name_for_match, level_1)
mr_us_new.drop(['genre','movie_name_for_match','level_1'],1, inplace=True)
```

```python
#modify the column names to make it clear
mr_us_new.rename(columns={'n_of_votes':'n_of_votes_imdb'},inplace=True)
```

```python
mr_us_new.head()
```

```python
mr_us_new.to_csv('mr_us_new.csv')
```
