```python
df_rp1=df1[['imdb_rating','meta_rating','n_of_votes']]
```

```python
group4.size().plot.pie(figsize=(12,12),textprops={'fontsize': 14})
plt.savefig('pie_inter.png',dpi=300)
```

```python
#boxplots for imdb_rating vs meta_rating and imdb_rating vs douban_rating
#but we need to arrange data frame structure first
df_dic1={'type':[],'rating':[]}
for rating in df1.imdb_rating:
    df_dic1['type']=df_dic1['type']+['imdb_rating']
    df_dic1['rating']=df_dic1['rating']+[rating]
    
for rating in df1.meta_rating:
    df_dic1['type']=df_dic1['type']+['meta_rating']
    df_dic1['rating']=df_dic1['rating']+[rating]
df_dic1
```

```python
df_box1=pd.DataFrame(df_dic1)
```

```python
print(df_box1.shape)
df_box1.head()
```

```python
df_dic2={'type':[],'rating':[]}
for rating in df2.imdb_rating:
    df_dic2['type']=df_dic2['type']+['imdb_rating']
    df_dic2['rating']=df_dic2['rating']+[rating]
    
for rating in df2.douban_rating:
    df_dic2['type']=df_dic2['type']+['douban_rating']
    df_dic2['rating']=df_dic2['rating']+[rating]
df_dic2
```

```python
df_box2=pd.DataFrame(df_dic2)
```

```python
print(df_box2.shape)
df_box2.head()
```
